### Transformación de variables numéricas

#### Codificación de un paso (one-hot encoding)

```python
df['Owner_Type'].unique()
```

```python
pd.get_dummies(df,columns=['Owner_Type'])
```

#### Codificación de etiquetas (label encoding)


```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(df['Owner_Type'])
df['Owner_Type_le'] = le.transform(df['Owner_Type'])
df.head()
```

#### Feature hash


```python
from sklearn.feature_extraction import FeatureHasher
```

```python
fh = FeatureHasher(n_features=3, input_type='string')
hashed = fh.transform(df[['Owner_Type']].astype(str).values)
hashed = pd.DataFrame(hashed.todense())
hashed.head()
```

```python
hashed.columns = ['Owner_Type_fh' + str(i) for i in hashed.columns]
pd.concat([df, hashed], axis=1)
```

#### Codificación binaria

```python
from category_encoders import BinaryEncoder


be = BinaryEncoder()
df_be = be.fit_transform(df['Owner_Type'])
pd.concat([df,df_be],axis=1)
```

#### Codificación Base N

```python
from category_encoders import BaseNEncoder

bne = BaseNEncoder(base=3)
df_bne = bne.fit_transform(df['Owner_Type'])
pd.concat([df, df_bne],axis=1)
```

#### Codificación de frecuencias


```python
df_train = pd.read_csv('C:/Users/HP/Downloads/train3.csv')
df_test = pd.read_csv('C:/Users/HP/Downloads/test3.csv')
```

```python
freq = df_train['Owner_Type'].value_counts()
df_train['Owner_Type_fe'] = df_train['Owner_Type'].map(freq)
df_test['Owner_Type_fe'] = df_test['Owner_Type'].map(freq)
```
