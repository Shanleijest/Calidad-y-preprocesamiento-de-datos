### Transformación de variables numéricas

#### Codificación de un paso (one-hot encoding)

```python
df['original_language'].unique()
```

```python
pd.get_dummies(df,columns=['original_language'])
```

#### Codificación de etiquetas (label encoding)


```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(df['original_language'])
df['original_language_le'] = le.transform(df['original_language'])
df.head()
```

#### Feature hash


```python
from sklearn.feature_extraction import FeatureHasher
```

```python
fh = FeatureHasher(n_features=3, input_type='string')
hashed = fh.transform(df[['original_language']].astype(str).values)
hashed = pd.DataFrame(hashed.todense())
hashed.head()
```

```python
hashed.columns = ['original_language_fh' + str(i) for i in hashed.columns]
pd.concat([df, hashed], axis=1)
```

#### Codificación binaria

```python
from category_encoders import BinaryEncoder


be = BinaryEncoder()
df_be = be.fit_transform(df['original_language'])
pd.concat([df,df_be],axis=1)
```

#### Codificación Base N

```python
from category_encoders import BaseNEncoder

bne = BaseNEncoder(base=3)
df_bne = bne.fit_transform(df['original_language'])
pd.concat([df, df_bne],axis=1)
```

#### Codificación de frecuencias


```python
df_train = pd.read_csv('C:/Users/HP/Downloads/train2.csv')
df_test = pd.read_csv('C:/Users/HP/Downloads/test2.csv')
```

```python
freq = df_train['original_language'].value_counts()
df_train['original_language_fe'] = df_train['original_language'].map(freq)
df_test['original_language_fe'] = df_test['original_language'].map(freq)
```
